```
# import libraries using pip/conda
import cv2
import os
import time
import uuid
```

```
# provide the path where you want to save the image
Images_Path = "drive/My Drive/Two-Way Comm/Implementation/SampleImages/CollectedImages"

# to cross check whether the actual path exist or not
if Images_Path:
  print("succesful")
```

```
#As were testing it, only few sign were recorderd lets say 5
#I would like to gather each sign in respective folder, so we create those particular labels

labels = ['Hello', 'Thanks', 'Yes', 'No', 'I love you']
num_images = 10 # this is the no of frames were recording from the video
# as the signs are static in nature we can simply use 10 or more, would be sufficient
```

```
#Now its time for to record the video and capture frames from it
for label in labels: # we should do this for each label
  !mkdir{"drive\My Drive\Two-Way Comm\Implementation\SampleImages\CollectedImages\\"+label} #it looks for the label in the specified as it not available it creates one and updates the images in respective label folder
  cvi = cv2.VideoCapture(0) # here were creating an instance of opencv, whcih enables us to capture the frames, from the camera feed
  print("Collectin images for {}".format(label))
  time.sleep(5)
  for num in range(num_images):
    flag, frame = cvi.read()
    imagename = os.path.join(Images_Path, label, label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
    cv2.imwrite(imagename, frame)
    cv2.show('frame', frame)
    time.sleep(2)


    if cv2.waitKey(1) & 0xFF == ord('q'):
      break
  
  cvi.release()
```
